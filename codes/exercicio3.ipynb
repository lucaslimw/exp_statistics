## A. Planeje um experimento na sua área de atuação no delineamento inteiramente casualizado.

```python
import pandas as pd

dados = pd.read_csv("G:/Meu Drive/Doutorado_Esalq/estatistica_exp/arquivos/dados.csv", sep=";")
```

O banco de dados fictício é referente a uma coleta de dados de vazão, em três cursos hidrícos distintos do rio Piracicaba. São três estações sendo realizadas 12 repetições em cada estação ao longo de um dia.
- Tratamento: **3 estações**;
- Resposta: **Vazão observada** na estação.


```python
print(dados)
```

## B. Obtenha um conjunto de dados da sua área, coletado num experimento instalado no delineamento inteiramente casualizado, faça a análise da variância e interprete os resultados.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Ajustando o modelo
modelo = ols('vazao ~ estacao', data=dados).fit()

# ANOVA
anova_tabela = sm.stats.anova_lm(modelo, typ=2)

print(anova_tabela)
```

A partir da análise de variância podemos concluir que não há diferença significativa entre as vazões das estações de coleta de água em diferentes cursos hídricos, ou seja, aceitamos H0 de que as médias dos grupos analisados são todas iguais.
