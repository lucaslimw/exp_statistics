# 1) Verificação da ortogonalidade entre os contrastes

Os contrastes fornecidos são:

- Y1 = mu1 - mu2  
- Y2 = 2 * mu2 - mu3 - mu4  
- Y3 = mu3 - mu4

## Vetores de coeficientes

Representamos cada contraste por um vetor de coeficientes:

- c1 = [1, -1, 0, 0]  
- c2 = [0, 2, -1, -1]  
- c3 = [0, 0, 1, -1]

## Produto escalar entre os pares

Vamos calcular o produto escalar entre os pares de vetores:

- c1 ⋅ c2 = (1*0) + (-1*2) + (0*-1) + (0*-1) = -2 → **Não ortogonais**  
- c1 ⋅ c3 = (1*0) + (-1*0) + (0*1) + (0*-1) = 0 → **Ortogonais**  
- c2 ⋅ c3 = (0*0) + (2*0) + (-1*1) + (-1*-1) = 0 → **Ortogonais**

## Conclusão

- Y1 ⟂ Y3  
- Y2 ⟂ Y3  
- Y1 não é ortogonal a Y2


# 2) Construção de um grupo de contrastes ortogonais

**Contexto do experimento:**

Estamos avaliando o efeito de quatro tipos de uso do solo sobre a vazão média mensal em microbacias da Bacia dos Três Unidos. Os tratamentos são:

- mu1: Floresta nativa  
- mu2: Agricultura convencional  
- mu3: Agricultura com plantio direto  
- mu4: Pastagem degradada  

Nosso objetivo é criar três contrastes ortogonais para comparar os efeitos dos usos do solo sobre a vazão.

---

**Construção dos contrastes:**

1. **Contraste 1 — Conservacionistas vs Intensivos**  
   Compara floresta + plantio direto com agricultura convencional + pastagem.  
   Fórmula: (mu1 + mu3) - (mu2 + mu4)  
   Coeficientes: [1, -1, 1, -1]

2. **Contraste 2 — Floresta vs Plantio Direto**  
   Compara as duas práticas conservacionistas.  
   Fórmula: mu1 - mu3  
   Coeficientes: [1, 0, -1, 0]

3. **Contraste 3 — Agricultura Convencional vs Pastagem**  
   Compara os usos mais intensivos.  
   Fórmula: mu2 - mu4  
   Coeficientes: [0, 1, 0, -1]

---

**Verificação da ortogonalidade:**


- Contraste 1 x Contraste 2: (1)(1) + (-1)(0) + (1)(-1) + (-1)(0) = 0  
- Contraste 1 x Contraste 3: (1)(0) + (-1)(1) + (1)(0) + (-1)(-1) = 0  
- Contraste 2 x Contraste 3: (1)(0) + (0)(1) + (-1)(0) + (0)(-1) = 0  

Todos as comparações deram zero, então os contrastes são ortogonais entre si.

---

**Conclusão:**

Grupo de contrastes ortogonais:

1. mu1 + mu3 - mu2 - mu4  
2. mu1 - mu3  
3. mu2 - mu4  

Todos são contrastes ortogonais.
